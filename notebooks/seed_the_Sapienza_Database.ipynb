```python
from daos.experiment_dao import ExperimentDao
import os
import time as tm
```

```python
experimentDao = ExperimentDao()

data_dir = "/home/michele/data/biassiale/Experiments"
print("SEEDING SAPIENZA DATABASE WITH EXISTING TDMS FILES:")
total_time = tm.time()
no_extension_files = []
for experiment_folder_name in sorted(os.listdir(data_dir)):
    experiment_folder_path = os.path.join(data_dir,experiment_folder_name)
    for file_name in os.listdir(experiment_folder_path):
        try:
            experiment_name, extension = file_name.split(".")
            if extension == "tdms":
                    file_path = os.path.join(experiment_folder_path, file_name)

                    print(f"\tPROCESSING DATA IN:  {file_name}\t")
                    start_time = tm.time()
                    experimentDao.create_experiment_from_file(file_path=file_path)
                    print(f"\t---Processing Time: {tm.time()-start_time}---")
        except:
            no_extension_files.append(file_name)
print(f"---TOTAL TIME TO UPLOAD SAPIENZA EXPERIMENTS: {tm.time()-total_time}---")
```
